```python
import matplotlib.pyplot as plt 

df = pd.read_hdf('./results.hdf','cancer_kmeans')

fig = plt.figure()
ax = df.plot(y='SSE', title="Cancer - Kmeans - SSE vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="SSE")
plt.savefig('./Plots/cancer_k_sse.png')

fig = plt.figure()
ax = df.plot(y='ARI', title="Cancer - Kmeans - ARI vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="ARI")
plt.savefig('./Plots/cancer_k_ari.png')

fig = plt.figure()
ax = df.plot(y='AMI', title="Cancer - Kmeans - AMI vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="AMI")
plt.savefig('./Plots/cancer_k_ami.png')

fig = plt.figure()
ax = df.plot(y='HOM', title="Cancer - Kmeans - Homogeneity vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="Homogeneity")
plt.savefig('./Plots/cancer_k_homo.png')

fig = plt.figure()
ax = df.plot(y='COM', title="Cancer - Kmeans - Completeness vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="Completeness")
plt.savefig('./Plots/cancer_k_comp.png')

fig = plt.figure()
ax = df.plot(y='SIL', title="Cancer - Kmeans - Silhouette Score vs Number of Clusters")
ax.set(xlabel="Number of Clusters", ylabel="Silhouette Score")
plt.savefig('./Plots/cancer_k_silh.png')
```
